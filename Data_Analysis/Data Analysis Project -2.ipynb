# Rolling Window Calculation:

```python
# Calculate a rolling mean for 'Mischief'
df['Rolling Mischief']=df['Mischief'].rolling(window=3).mean()
df
```

```python
# Calculate a rolling sum for 'Mischief'

df['Rolling Mischief Sum']=df['Mischief'].rolling(window=3).sum()
df
```

```python
# Calculate Cumulative Product for 'Cuteness'

df['Cumulative Product of Cuteness']=df['Cuteness'].cumprod()
df
```

```python
# Calculate Cumulative sum for 'Cuteness'

df['Cumulative Cuteness']=df['Cuteness'].cumsum()
df
```
